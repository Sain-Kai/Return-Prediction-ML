```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
from sklearn.preprocessing import LabelEncoder
from xgboost import XGBClassifier
```

```python
db = pd.read_excel("Retail-Supply-Chain-Sales-Dataset.xlsx")
```

```python
db.head()
```

```python
db.info()
```

```python
target = db.iloc[:,[4,6,7,9] + list(range(17,22))].copy()
```

```python
target.sample(200)
```

```python
target['Duration'] =db['Ship Date']-db['Order Date']
```

```python
target.info()
```

```python
target.sample(20)
```

```python
target['Returned']=db['Returned'].map({'Yes':1,'Not':0})
```

```python
target.sample(10)
```

```python
encoding_cols = ['Ship Mode','Segment','City','Product Name']

for cols in encoding_cols:
    le=LabelEncoder()
    le.fit_transform(target[cols])
    target[cols]=le.transform(target[cols])
```
